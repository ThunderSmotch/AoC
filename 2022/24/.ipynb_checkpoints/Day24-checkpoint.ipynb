What is the fewest number of minutes required to avoid the blizzards and reach the goal?


```python
import numpy as np
lines = open("input.txt").read().splitlines()

height = len(lines) - 2
width = len(lines[0]) - 2

start = 0-1j
end   = (width - 1) + height*1j

blizzards = []

j = -1
for line in lines:
  i = -1
  for char in line:
    if char == ">":
      blizzards.append((i+j*1j, 0))
    elif char == "v":
      blizzards.append((i+j*1j, 1))
    elif char == "<":
      blizzards.append((i+j*1j, 2))
    elif char == "^":
      blizzards.append((i+j*1j, 3))

    i += 1
  j += 1

total_cycles = np.lcm(width,height)

directions = [1, 1j, -1, -1j]

grid = {}
grid[0] = blizzards.copy()

for i in range(1, total_cycles):
  new_blizzards = []

  for blizz in grid[i-1]:
    pos, dirr = blizz
    pos += directions[dirr]
    if int(pos.real) == -1:
      pos += width
    elif int(pos.real) == width:
      pos -= width
    elif int(pos.imag) == -1:
      pos += height*1j
    elif int(pos.imag) == height:
      pos -= height*1j
    new_blizzards.append((pos, dirr))

  grid[i] = new_blizzards.copy()

blizzards = {}
for i in range(total_cycles):
  blizzards[i] = [pos for pos, _ in grid[i] ]

def travel_from_to(start, end, offset = 0):
  tree = {0: [start]}

  reached_end = False
  t = 0

  while(not reached_end):
    #print("After", t, "minutes, we have a total of ", len(tree[t]), "possible positions")
    states = []
    for state in tree[t]:
      if not state in blizzards[(t+1 + offset) % total_cycles]:
        states.append(state) # Stay put
      
      for direction in directions: # Try to move
        new_pos = state + direction

        if new_pos == end:
          print(state, t)
          states.append(new_pos)
          reached_end = True
          break
        
        if 0 <= new_pos.real <= width - 1 and 0 <= new_pos.imag <= height - 1:
          if not new_pos in blizzards[(t+1 + offset) % total_cycles]:
            if not new_pos in states:
              states.append(new_pos)
    t += 1
    tree[t] = states

  print("Got to end in", t, "steps")
  return t
```

```python
travel_from_to(start, end)
```
