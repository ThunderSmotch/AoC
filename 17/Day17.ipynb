## --- Day 17: Pyroclastic Flow ---

Your handheld device has located an alternative exit from the cave for you and the elephants. The ground is rumbling almost continuously now, but the strange valves bought you some time. It's definitely getting warmer in here, though.

The tunnels eventually open into a very tall, narrow chamber. Large, oddly-shaped rocks are falling into the chamber from above, presumably due to all the rumbling. If you can't work out where the rocks will fall next, you might be crushed!


The rocks fall in the order shown above: first the - shape, then the + shape, and so on. Once the end of the list is reached, the same order repeats: the - shape falls first, sixth, 11th, 16th, etc.

The rocks don't spin, but they do get pushed around by jets of hot gas coming out of the walls themselves. A quick scan reveals the effect the jets of hot gas will have on the rocks as they fall (your puzzle input).

For example, suppose this was the jet pattern in your cave:
```
>>><<><>><<<>><>>><<<>>><<<><<<>><>><<>>
```
In jet patterns, < means a push to the left, while > means a push to the right. The pattern above means that the jets will push a falling rock right, then right, then right, then left, then left, then right, and so on. If the end of the list is reached, it repeats.

The tall, vertical chamber is exactly seven units wide. Each rock appears so that its left edge is two units away from the left wall and its bottom edge is three units above the highest rock in the room (or the floor, if there isn't one).

After a rock appears, it alternates between being pushed by a jet of hot gas one unit (in the direction indicated by the next symbol in the jet pattern) and then falling one unit down. If any movement would cause any part of the rock to move into the walls, floor, or a stopped rock, the movement instead does not occur. If a downward movement would have caused a falling rock to move into the floor or an already-fallen rock, the falling rock stops where it is (having landed on something) and a new rock immediately begins falling.

To prove to the elephants your simulation is accurate, they want to know how tall the tower will get after 2022 rocks have stopped (but before the 2023rd rock begins falling). In this example, the tower of rocks will be 3068 units tall.

How many units tall will the tower of rocks be after 2022 rocks have stopped falling?


````python
f = open("input.txt", "r")
pattern = f.read().splitlines()[0]

blocks = ["hline", "plus", "L", "vline", "square"]

block = {
    "hline": [[2,2,2,2]],
    "plus": [[0,2,0],
             [2,2,2],
             [0,2,0]],
    "L": [[2,2,2],
          [0,0,2],
          [0,0,2]],
    "vline": [[2],
              [2],
              [2],
              [2]],
    "square": [[2,2],
               [2,2]]
}

# Game Grid
# 0 is air, 1 is block, 2 is falling block
class Grid:
    def __init__(self, width = 7):
        self.width = width
        self.grid = {0: [1,1,1,1,1,1,1]}
        self.max_height = 0
    
    def __getitem__(self, y, x = None):
        if not y in self.grid.keys():
            self.grid[y] = [0,0,0,0,0,0,0]        
        if x == None:
            return self.grid[y]
        else:
            return self.grid[y][x]
    
    def draw(self, ymin, ymax):
        for y in range(ymax, ymin - 1, -1):
            string = ""
            for pixel in self[y]:
                if pixel == 0:
                    string += "."
                elif pixel == 1:
                    string += "@"
                else:
                    string += "#"
            print(string)

# Coordinates are relative to the bottom-left pixel of the block
class Block:
    def __init__(self, y, x, block_type):
        self.x = x
        self.y = y
        
        self.block_type = block_type
        self.pixels = block[block_type]
        
        self.w = len(self.pixels[0])
        self.h = len(self.pixels)
        
    def is_valid_position(self, x, y):       
        
        # Check if it fits inside pit's walls
        if x < 0 or x + self.w - 1 > 6:
            return False
        
        # Check if the block pixels do not intersect existing pit blocks
        for yy in range(self.h):
            for xx in range(self.w):
                if self.pixels[yy][xx] == 2 and grid[y + yy][x + xx] == 1:
                    return False
        
        return True
    
    def move(self, dx=0, dy=0):
        if self.is_valid_position(self.x + dx, self.y + dy):
            self.x += dx
            self.y += dy
            return True
        return False
    
    def stop(self):
        for yy in range(self.h):
            for xx in range(self.w):
                if self.pixels[yy][xx] == 2:
                    grid[self.y + yy][self.x + xx] = 1
                    
        if self.h + self.y - 1 > grid.max_height:
            grid.max_height = self.h + self.y - 1
        
def tetris(total_rocks = 2022, start_block_index = 0, start_pattern_index = 0, find_pidx = 6, make_counts = False):
    # GAME START
    global grid
    grid = Grid()
    fallen_rocks = 0

    pattern_length = len(pattern)
    blocktype_length = len(blocks)

    b_idx = start_block_index
    p_idx = start_pattern_index

    # part2
    counter = 0
    counts = [0]*pattern_length
    
    while fallen_rocks < total_rocks:
        # GENERATE BLOCK
        rock = Block(grid.max_height + 4, 2, blocks[b_idx])
        while True:
            # JET
            if pattern[p_idx] == ">":
                rock.move(dx = 1)
            else:
                rock.move(dx = -1)
            p_idx = 0 if p_idx == pattern_length - 1 else p_idx + 1

            # FALL
            if not rock.move(dy = -1):
                break
        # Stop rock
        rock.stop()
        fallen_rocks += 1

        if b_idx == 0 and make_counts:
            counts[p_idx] += 1
        
        if b_idx == 0 and p_idx == find_pidx and counter < 3:
            print(fallen_rocks, grid.max_height)
            grid.draw(ymin=grid.max_height - 10, ymax = grid.max_height + 4)
            counter += 1

        b_idx = 0 if b_idx == blocktype_length - 1 else b_idx + 1
    
    print("Final max height:", grid.max_height)        
    
    if make_counts:
        for count in counts:
            if count > 3:
                print("Found repeat at p_idx", counts.index(count))
                break
        
    return grid.max_height

tetris(10000, make_counts=False, find_pidx=35)
grid.max_height
````

# --- Part Two ---

The elephants are not impressed by your simulation. They demand to know how tall the tower will be after 1000000000000 rocks have stopped! Only then will they feel confident enough to proceed through the cave.

In the example above, the tower would be 1514285714288 units tall!

How tall will the tower be after 1000000000000 rocks have stopped?


````python
# my strategy was to find a periodic sequence inside the tetris game!
# once i found a loop then i just simulate it the needed amount of times and then simulate the remaining blocks O_O

# a more automated approach could just save the periodic table of changes in max_height and use that to reach the correct answer
````

````python
# For the test case
a1, h1 = 86, 132
a2, h2 = 121, 185
rocks_fallen  = a2-a1
height_gained = h2-h1
total_rocks = 1000000000000

h1 + ((total_rocks - a1)//rocks_fallen)*height_gained + tetris(total_rocks = (total_rocks-a1)%rocks_fallen, start_block_index=1, start_pattern_index=6)
````
