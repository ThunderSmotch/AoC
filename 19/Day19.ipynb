However, by using blueprint 2 in the example above, you could do even better: the largest number of geodes you could open in 24 minutes is 12.

Determine the quality level of each blueprint by multiplying that blueprint's ID number with the largest number of geodes that can be opened in 24 minutes using that blueprint. In this example, the first blueprint has ID 1 and can open 9 geodes, so its quality level is 9. The second blueprint has ID 2 and can open 12 geodes, so its quality level is 24. Finally, if you add up the quality levels of all of the blueprints in the list, you get 33.

Determine the quality level of each blueprint using the largest number of geodes it could produce in 24 minutes. What do you get if you add up the quality level of all of the blueprints in your list?


````python
import numpy as np

f = open("input.txt", "r")
lines = f.read().splitlines()

# Parsing input
blueprints = []
for line in lines:
    words = line.split("costs ")
    blueprints.append({
        "bid": int(words[0][words[0].index(" ") + 1:words[0].index(":")]),
        "ore_robot": np.array([int(words[1][:words[1].index(" ")]), 0, 0, 0]),
        "clay_robot": np.array([int(words[2][:words[2].index(" ")]), 0, 0, 0]),
        "obsidian_robot": np.array([int(words[3][:words[3].index(" ")]), int(words[3][words[3].index("and ") + 4:words[3].index(" clay")]), 0, 0]),
        "geode_robot": np.array([int(words[4][:words[4].index(" ")]), 0, int(words[4][words[4].index("and ") + 4:words[4].index(" obsidian")]), 0])
    })

geodes_array = []
i = 0
#27 -> 6
# 6 -> 1

# Find max geodes for each blueprint
for blueprint in blueprints[:1]:
    i += 1
    print("Blueprint", i)
    
    total_time = 24

    # ore clay obsidian geode
    robots = np.array([1,0,0,0])
    resources = np.array([0,0,0,0])
    
    tree = {}
    for t in range(total_time + 1):
        tree[t] = []
    tree[0].append((robots, resources))



    time = 0

    # Pruning flags
    geo_reached = False
    geo_t = -1
    geo_max = 0
    geo_max_t = -1

    obs_reached = False
    obs_t = -1
    obs_max = 0
    obs_max_t = -1

    max_score = -1
    old_max_score = -1
    prop = 0.8
    
    max_ore_robots = max(blueprint["geode_robot"][0], blueprint["obsidian_robot"][0], blueprint["clay_robot"][0], blueprint["ore_robot"][0])
    max_clay_robots = max(blueprint["geode_robot"][1], blueprint["obsidian_robot"][1], blueprint["clay_robot"][1], blueprint["ore_robot"][1])
    
    while time < total_time:
        print(time, len(tree[time]))
        
        for state in tree[time]: # branch out from each possible state
            
            robots, resources = state
            
            if geo_reached and time > geo_t + 1 and robots[3] == 0:
                continue
            if obs_reached and time > obs_t + 2 and robots[2] == 0:
                continue
            if time >= geo_max_t and robots[3] < geo_max - 2:
                continue
            if not geo_reached and time >= obs_max_t and robots[2] < obs_max - 2:
                continue

            future_resources = resources + robots
            
            # If can build geode then just do it as that is ALWAYS the best
            if (blueprint["geode_robot"] <= resources).all():  
                tree[time + 1].append( (robots + np.array([0,0,0,1]), future_resources - blueprint["geode_robot"]) )

                if robots[3] + 1 > geo_max:
                    geo_max = robots[3] + 1
                    geo_max_t = time + 1

                if not geo_reached:
                    geo_reached = True
                    geo_t = time + 1
                
                continue

            # If can build obsidian then just do it
            if (blueprint["obsidian_robot"] <= resources).all():  # If can build obsidian robot            
                tree[time + 1].append( (robots + np.array([0,0,1,0]), future_resources - blueprint["obsidian_robot"]) )

                if robots[2] + 1 > obs_max and not geo_reached:
                    obs_max = robots[2] + 1
                    obs_max_t = time + 1
                
                if not obs_reached:
                    obs_reached = True
                    obs_t = time + 1
                
                continue
            
            # Else do nothing, or build ore or clay
            tree[time + 1].append( (robots, future_resources) ) # Do nothing
            if (blueprint["ore_robot"] <= resources).all() and robots[0] < max_ore_robots:  # If can build ore robot            
                tree[time + 1].append( (robots + np.array([1,0,0,0]), future_resources - blueprint["ore_robot"]) )
            if (blueprint["clay_robot"] <= resources).all() and robots[1] < max_clay_robots:  # If can build clay robot            
                tree[time + 1].append( (robots + np.array([0,1,0,0]), future_resources - blueprint["clay_robot"]) )                
        
        time += 1

    max_geodes = 0
    for state in tree[time]:
        max_geodes = max(max_geodes, state[1][3])
    geodes_array.append(max_geodes)

total = 0
for i in range(len(geodes_array)):
    total += geodes_array[i]*(i+1)
print(geodes_array)
total
````

````python
MAX_QUEUE_SIZE = 1000000
def solve(blueprint, total_time = 24):
    max_ore_robots = max(blueprint["geode_robot"][0], blueprint["obsidian_robot"][0], blueprint["clay_robot"][0], blueprint["ore_robot"][0])
    max_clay_robots = max(blueprint["geode_robot"][1], blueprint["obsidian_robot"][1], blueprint["clay_robot"][1], blueprint["ore_robot"][1])

    
    # ore clay obsidian geode
    time = 0
    robots = np.array([1,0,0,0])
    resources = np.array([0,0,0,0])
    
    queue = deque()
    seen = set()
    
    queue.append((0, robots, resources))

    maxG = 0
    
    while queue:
        
        if len(queue) > MAX_QUEUE_SIZE:
            print("Reached max queue size!!!")
        
        t, robots, resources = queue.popleft()
        
        future_resources = resources + robots
    
        if t == total_time:
            maxG = max(maxG, resources[3])
            continue
            
        if (t, tuple(robots), tuple(resources)) in seen:
            continue
        seen.add((t, tuple(robots), tuple(resources)))
    
        # If can build geode then just do it as that is ALWAYS the best
        if (blueprint["geode_robot"] <= resources).all():  
            queue.appendleft( (t+1, robots + np.array([0,0,0,1]), future_resources - blueprint["geode_robot"]) )
            continue

        # If can build obsidian then just do it
        if (blueprint["obsidian_robot"] <= resources).all():  # If can build obsidian robot            
            queue.appendleft( (t+1, robots + np.array([0,0,1,0]), future_resources - blueprint["obsidian_robot"]) )
            continue

        # Else do nothing, or build ore or clay
        queue.appendleft( (t+1, robots, future_resources) ) # Do nothing
        if (blueprint["ore_robot"] <= resources).all() and robots[0] < max_ore_robots:  # If can build ore robot            
            queue.appendleft( (t+1, robots + np.array([1,0,0,0]), future_resources - blueprint["ore_robot"]) )
        if (blueprint["clay_robot"] <= resources).all() and robots[1] < max_clay_robots:  # If can build clay robot            
            queue.appendleft( (t+1, robots + np.array([0,1,0,0]), future_resources - blueprint["clay_robot"]) )
    
    return maxG
````

````python
# DFS search
from collections import deque
geodes_array = []
i = 0

# Find max geodes for each blueprint
for blueprint in blueprints:
    i += 1
    print("Blueprint", i)
    geodes_array.append(solve(blueprint))

total = 0
for i in range(len(geodes_array)):
    total += geodes_array[i]*(i+1)
print(geodes_array)
total
````
