However, by using blueprint 2 in the example above, you could do even better: the largest number of geodes you could open in 24 minutes is 12.

Determine the quality level of each blueprint by multiplying that blueprint's ID number with the largest number of geodes that can be opened in 24 minutes using that blueprint. In this example, the first blueprint has ID 1 and can open 9 geodes, so its quality level is 9. The second blueprint has ID 2 and can open 12 geodes, so its quality level is 24. Finally, if you add up the quality levels of all of the blueprints in the list, you get 33.

Determine the quality level of each blueprint using the largest number of geodes it could produce in 24 minutes. What do you get if you add up the quality level of all of the blueprints in your list?


````python
from math import ceil

f = open("input_test.txt", "r")
lines = f.read().splitlines()

blueprints = []
for line in lines:
    words = line.split("costs ")
    blueprints.append({
        "bid": int(words[0][words[0].index(" ") + 1:words[0].index(":")]),
        "ore_robot_ore": int(words[1][:words[1].index(" ")]),
        "clay_robot_ore": int(words[2][:words[2].index(" ")]),
        "obsidian_robot_ore": int(words[3][:words[3].index(" ")]),
        "obsidian_robot_clay": int(words[3][words[3].index("and ") + 4:words[3].index(" clay")]),
        "geode_robot_ore": int(words[4][:words[4].index(" ")]),
        "geode_robot_obsidian": int(words[4][words[4].index("and ") + 4:words[4].index(" obsidian")]),
    })

total_time = 24
robots = {
    "ore": 1,
    "clay": 0,
    "obsidian": 0,
    "geode": 0
}
resources = {
    "ore": 0,
    "clay": 0,
    "obsidian": 0,
    "geode": 0
}

print(blueprints)

blueprint = blueprints[1]


constraints = {
    "ore": 100,
    "clay": 100,
    "obsidian": 100,
}

time = 0
robot = "clay"
while time < 24:
    
    old_robots = robots.copy()
    
    # Production of robots
    if robot == "ore" and robots["ore"] < constraints["ore"]:
        if blueprint["ore_robot_ore"] <= resources["ore"]:
            resources["ore"] -= blueprint["ore_robot_ore"]
            robots["ore"] += 1
    elif robot == "clay" and robots["clay"] < constraints["clay"]:
        if blueprint["clay_robot_ore"] <= resources["ore"]:
            resources["ore"] -= blueprint["clay_robot_ore"]
            robots["clay"] += 1
    elif robot == "obsidian" and robots["obsidian"] < constraints["obsidian"]:
        if blueprint["obsidian_robot_clay"] <= resources["clay"] and blueprint["obsidian_robot_ore"] <= resources["ore"]:
            resources["ore"] -= blueprint["obsidian_robot_ore"]
            resources["clay"] -= blueprint["obsidian_robot_clay"]
            robots["obsidian"] += 1
    elif robot == "geode":
        if blueprint["geode_robot_obsidian"] <= resources["obsidian"] and blueprint["geode_robot_ore"] <= resources["ore"]:
            resources["ore"] -= blueprint["geode_robot_ore"]
            resources["obsidian"] -= blueprint["geode_robot_obsidian"]
            robots["geode"] += 1
    
    # Gathering resources
    for resource in resources.keys():
        resources[resource] += old_robots[resource]
    
    # Heuristic to decide next robot to build
    priority = ["geode", "obsidian", "clay", "ore"]
    for rb in priority:
        if rb == "geode" and robots["obsidian"] > 0: # If we can build geode in the next turn then wait
            # Rounds left if nothing is done
            rl_obsidian = ceil( (blueprint["geode_robot_obsidian"] - resources["obsidian"]) / robots["obsidian"] )
            rl_ore = ceil( (blueprint["geode_robot_ore"] - resources["ore"]) / robots["ore"] )
            rl = max(rl_ore, rl_obsidian)
            
            # Rounds left if obsidian is built
            rlb_obsidian = ceil( (blueprint["geode_robot_obsidian"] - resources["obsidian"]) / (robots["obsidian"] + 1) )
            rlb_ore = ceil( (blueprint["geode_robot_ore"] + blueprint["obsidian_robot_ore"] - resources["ore"]) / robots["ore"] )
            rlb = max(rlb_obsidian, rlb_ore)
            
            print("Rounds to geode", rl)
            print("Rounds to geode if I make obsidian now", rlb)
            
            if rl <= rlb:
                robot = "geode"
                break            
        elif rb == "obsidian" and robots["clay"] > 0: # If we can build obsidian in the next few turns then wait
            # Rounds left if nothing is done
            rl_clay = ceil((blueprint["obsidian_robot_clay"] - resources["clay"])/robots["clay"])
            rl_ore = ceil( (blueprint["obsidian_robot_ore"] - resources["ore"] )/robots["ore"] )
            rl = max(rl_ore, rl_clay)
            
            # Rounds left if clay is built
            rlc_clay = ceil((blueprint["obsidian_robot_clay"] - resources["clay"])/(robots["clay"] + 1))
            rlc_ore = ceil( (blueprint["obsidian_robot_ore"] - resources["ore"] + blueprint["clay_robot_ore"] )/robots["ore"] )
            rlc = max(rlc_ore, rlc_clay)
            
            # Rounds left if ore is built
            rlo_ore = ceil( (blueprint["obsidian_robot_ore"] - resources["ore"] + blueprint["ore_robot_ore"])/(robots["ore"] + 1) )
            rlo = max(rl_clay, rlo_ore)
            
            # debug prints
            print("Rounds to obsidian", rl)
            print("Rounds to obsidian if I make clay now", rlc)
            print("Rounds to obsidian if I make ore now", rlo)
            
            if rl < rlc:
                robot = "obsidian"
                break
            elif rlo < rlc:
                robot = "clay"
                break
        elif rb == "clay": # Clay stage
            rl_ore = ceil( (blueprint["clay_robot_ore"] - resources["ore"])/robots["ore"] )
            rlo_ore = ceil( (blueprint["clay_robot_ore"] - resources["ore"] + blueprint["ore_robot_ore"])/(robots["ore"] + 1) )
            
            print("Rounds to clay", rl_ore)
            print("Rounds to clay if I make ore now", rlo_ore)
            
            if rlo_ore < rl_ore:
                robot = "ore"
            else:
                robot = "clay"
            break
    
    time += 1
    print("TIME", time, " - going to build", robot, " ====================")
    print("  >    Robots (", robots["ore"], robots["clay"], robots["obsidian"], robots["geode"] ,")")
    print("  > Resources (", resources["ore"], resources["clay"], resources["obsidian"], resources["geode"] ,")")

print("Blueprint", blueprint["bid"], "produced", resources["geode"], "geodes")
````
