Work out the steps to release the most pressure in 30 minutes. What is the most pressure you can release?

````python
f = open("input_test.txt", "r")
f = open("input.txt", "r")
#f = open("tt.txt", "r")
lines = f.read().splitlines()

valves = {}
positive_valves = []

for line in lines:
    words = line.split(" ")
    
    to = []
    for other_valve in words[9:]:
        to.append(other_valve[0:2])
    rate = int(words[4][5:-1])
    
    valves[words[1]] = {
        "rate": int(words[4][5:-1]),
        "to": to
    }
    
    if rate > 0:
        positive_valves.append(words[1])

distances = {}

for start_valve in valves.keys():
    
    visited = {
        
    }
    
    to_visit = []
    
    current_valve = (start_valve, 0)
    
    while True:
        # Add current to visited
        if not current_valve[0] in visited.keys() or current_valve[1] < visited[current_valve[0]]:
            visited[current_valve[0]] = current_valve[1]
        
        # Add unvisited neighbors
        neighbors = valves[current_valve[0]]["to"] # List of neighbors
        
        for neighbor in neighbors:
            if not neighbor in visited.keys():
                to_visit.append((neighbor, current_valve[1] + 1))
        
        if len(to_visit) == 0:
            break; # Visited all nodes
        else:
            current_valve = to_visit.pop(0)  
    
    # Set dictionary
    distances[start_valve] = visited
    
print(positive_valves)
valves
````

````python
%%time
start_valve = "AA"
no_positive = len(positive_valves)
def get_flow(state):
    flow = 0
    for i in range(no_positive):
        if state & (1 << i): # Valve is opened
            flow += valves[positive_valves[i]]["rate"]
    return flow

solutions = {}

# m is minutes remaining
# valve is current valve at this position
# state is bitset of the opened valves
def get_most_pressure(m, valve, state):
    if (m, valve, state) in solutions.keys(): # If it exists return it
        return solutions[(m, valve, state)]
    elif m == 1: # Any action is useless
        sol = (get_flow(state), "")
        solutions[(m, valve, state)] = sol
        return sol
    else:
        neighbors = [v for v in positive_valves if v != valve]      
        actions = []
        paths = []
        
        flow = get_flow(state)
        
        # Go to positive node    
        for neigh in neighbors:
            dist = distances[valve][neigh]
            remaining_minutes = m - dist
            if remaining_minutes > 0:
                t_flow, t_path = get_most_pressure(remaining_minutes, neigh, state)
                actions.append( t_flow + dist*flow )
                paths.append( t_path )
        
        # Or do nothing
        t_flow, t_path = get_most_pressure(1, valve, state)
        actions.append(t_flow + (m-1)*flow)
        paths.append(t_path)
        
        # Or open current valve
        if valve != start_valve:
            binary_valve = (1 << positive_valves.index(valve))
            if not binary_valve & state: # If this valve is not open
                t_flow, t_path = get_most_pressure(m - 1, valve, state | binary_valve)
                actions.append( t_flow + flow )
                paths.append(t_path + " " + valve)
                       
        sol = max(actions)
        path = paths[actions.index(sol)]
        
        solutions[(m, valve, state)] = (sol, path)
        return (sol, path)

get_most_pressure(30, "AA", 0)
````
