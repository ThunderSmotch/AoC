In decimal, the sum of these numbers is 4890.

As you go to input this number on Bob's console, you discover that some buttons you expected are missing. Instead, you are met with buttons labeled =, -, 0, 1, and 2. Bob needs the input value expressed as a SNAFU number, not in decimal.

Reversing the process, you can determine that for the decimal number 4890, the SNAFU number you need to supply to Bob's console is 2=-1=0.

The Elves are starting to get cold. What SNAFU number do you supply to Bob's console?

Your puzzle answer was 2==221=-002=0-02-000.

```python
lines = open("input.txt").read().splitlines()

symbols = {
    "=": -2,
    "-": -1,
    "0": 0,
    "1": 1,
    "2": 2,
}

def snafu_to_decimal(snafu):
  decimal = 0
  length = len(snafu)
  for i in range(length):
    decimal += (5**i)*symbols[snafu[length - 1 - i]]
  return decimal

def decimal_to_snafu(decimal):
  snafu = ""

  i = 0
  while(decimal >= 5**i):
      i+=1
  max_exponent = i-1

  # See if it starts with 1 or 2
  test_number = 5**max_exponent
  for i in range(0, max_exponent):
    test_number += 5**i

  if test_number <= decimal: # Starts with a two
    snafu += "2"
    for i in range(max_exponent):
      snafu += "=" 
    i = 1
    while(snafu_to_decimal(snafu) != decimal and i != max_exponent + 1):
      print(snafu)
      symbols = ["=", "-", "0", "1", "2"]
      for j in range(1, 5):
        snafu = snafu[:i] + symbols[j] + snafu[i+1:]
        if (snafu_to_decimal(snafu) > decimal):
          snafu = snafu[:i] + symbols[j-1] + snafu[i+1:]
          break
      i += 1
  else:                      # Starts with a one
    snafu += "1"
    for i in range(max_exponent):
      snafu += "2"
    print(snafu)
    i = 1
    while(snafu_to_decimal(snafu) != decimal and i != max_exponent + 1 ):
      symbols = ["2", "1", "0", "-", "="]
      for j in range(1, 5):
        snafu = snafu[:i] + symbols[j] + snafu[i+1:]
        if (snafu_to_decimal(snafu) < decimal):
          snafu = snafu[:i] + symbols[j-1] + snafu[i+1:]
          break
      i += 1

  return snafu

sum = 0
for line in lines:
  sum += snafu_to_decimal(line)
sum
```
